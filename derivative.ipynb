```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
def f(x):
    return x ** 5 + 4 * np.sin(2 * x) + np.cos(3 * x + 3)
x = 1
derivative(f, x, dx=1e-6, n=1), derivative(f, x, dx=1e-6, n=2)
```

Уровень 1
Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
def f(x):
    return (np.sin(2 * x + 1)) ** 5
x = np.linspace(-5, 5)
plt.plot(x, [derivative(f, i, dx=1e-6, n=1) for i in x])
plt.plot(x, [derivative(f, i, dx=1e-6, n=2) for i in x])
plt.show()
```
