```python
email = ['9m6@bk.ru', 'rubka07@bk.ru']
url_ = 'https://digibody.avast.com/v1/web/leaks'
df = pd.DataFrame()
for email_ in range(len(email)):
    name['email'] = email[email_]
    check = json.loads(requests.post(url_, data=json.dumps(name)).text)
    for x in range(len(check['value'])):
        df = pd.concat([df, pd.DataFrame([{'почта': name['email'],\
           'дата утечки': check['value'][x]['leak_info']['date'],\
           'источник утечки': check['value'][x]['leak_info']['title'],\
           'описание утечки': check['value'][x]['leak_info']['description']}])]) 
df
```
