Разделить данные на тестовую и обучающую выборки. 
Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary).
Оценить качество модели на тестовой выборки.

```python
X_train, X_test, y_train, y_test = train_test_split(df[['satisfaction_level', 'last_evaluation', 'number_project',
       'average_montly_hours', 'time_spend_company', 'Work_accident', 
       'promotion_last_5years']], df['left'].to_numpy(), test_size=0.25)
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
```

```python
lda.predict(X_test)
```

```python
result = pd.DataFrame([y_test, lda.predict(X_test)]).T
result
```

```python
f'Точность предсказания увольнения состовляет {accuracy_score(y_test, lda.predict(X_test))*100}%'
```
